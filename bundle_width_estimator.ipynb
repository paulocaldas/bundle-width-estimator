### notes about this script

function above estimates the aparent width of bundle filaments from fluorescence images. For a given movie as input, images are binarized using an adaptive threshold (corrects for non-homogeneous background intensities) and their background is removed using a block size of 21 pixels. The binarized time lapse movie is then processed with a denoise function to remove small particles and the Euclidean Distance Map (EDM) of every frame is estimated. This transformation results in a grey scale image, where the grey value of each pixel represents the shortest distance to the nearest pixel in the background. Accordingly, bundle widths correspond to the local peak intensities multiplied by 2. The mean bundle width for every frame of the movie is calculated by identifying the peak intensities for each line and column of the image. The output is an histogram for each frame showing the distribution of widths in time. In addition, the mean value of this distributons is plotted over time. For more details, check the main functions in the background (based on `skimage` library)
